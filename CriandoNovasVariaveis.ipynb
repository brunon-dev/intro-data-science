## Criando Novas Variáveis

```python
import pandas as pd
```

```python
dados = pd.read_csv("Aluguel_Residencial3.csv", sep = ";")
dados.head(10)
```

```python
# cria nova variável de Valor Bruto => Valor + Condominio + IPTU
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
dados.head(10)
```

```python
# cria nova variável de valor do metro quadrado => Valor / Area
dados['Valor M2'] = dados['Valor'] / dados['Area']
dados.head(10)
```

```python
# arredondando o valor do metro quadrado para apenas 2 casas decimais
dados['Valor M2'] = dados['Valor M2'].round(2)
dados.head(10)
```

```python
# cria também o valor bruto do metro quadrado
dados['Valor Bruto M2'] = (dados['Valor Bruto'] / dados['Area']).round(2)
dados.head(10)
```

```python
dados['Tipo'].unique()
```

```python
# cria uma variável que vai agregar todos os tipos de Casas (Casa, Casa de Condomínio, Casa de Vila) e classificar todo o resto como apartamento
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
dados.head(10)
```

## Excluindo Variáveis de um DataFrame

```python
dados_aux = pd.DataFrame(dados[['Tipo Agregado', 'Valor M2', 'Valor Bruto', 'Valor Bruto M2']])
dados_aux.head(10)
```

```python
# primeira forma de excluir uma variável
del dados_aux['Valor Bruto']
dados_aux.head(10)
```

```python
# segunda forma de excluir uma variável
dados_aux.pop('Valor Bruto M2')
```

```python
dados_aux.head(10)
```

```python
# terceira forma de excluir 1 ou mais variáveis - axis define o eixo do drop - axis = 1 é para coluna e axis = 0 é para linha
dados.drop(['Valor Bruto', 'Valor Bruto M2'], axis = 1, inplace = True)
dados.head(10)
```

```python
dados.to_csv('Aluguel_Residencial4.csv', sep = ";", index = False)
```
